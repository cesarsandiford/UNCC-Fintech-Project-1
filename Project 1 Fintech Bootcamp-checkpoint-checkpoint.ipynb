#### Study of different assets perfomance during Covid-19 Pandemic

```python
import os
import pandas as pd
import numpy as np
import empyrical
import hvplot.pandas
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
from MCForecastTools import MCSimulation

%matplotlib inline
```

```python
# List of type of assets in ETFs
# SP500(SPY), Bonds(AGG), Gold(GLD), Real Estate(USRT)
```

```python
load_dotenv()
```

```python
# Setting the variables for the Alpaca API and secret keys
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2"
)

# Setting the tickers for the portfolio
sp500 = ["SPY"]
bonds =["AGG"]
gold = ["GLD"]
reit = ["USRT"]

# Setting the timeframe to 1Day
timeframe ="1Day"

# Date as ISO format. Setting the start date a year prior to Covid-19 and end date at a year after.
start_date = pd.Timestamp("2019-03-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-03-01", tz="America/New_York").isoformat()
```

```python
# Use the Alpaca get_bars function to get current closing prices of each asset ETF.
usrt_df = alpaca.get_bars(
    reit,
    timeframe,
    start = start_date,
    end = end_date
).df

spy_df = alpaca.get_bars(
    sp500,
    timeframe,
    start = start_date,
    end = end_date
).df

agg_df = alpaca.get_bars(
    bonds,
    timeframe,
    start = start_date,
    end = end_date
).df

gld_df = alpaca.get_bars(
    gold,
    timeframe,
    start = start_date,
    end = end_date
).df


# Reviewing the first 5 rows 
display(usrt_df.head())
display(spy_df.head())
display(agg_df.head())
display(gld_df.head())
```

```python
# Setting the Daily Returns of each asset ETF.
close_sp500_df = pd.DataFrame()
close_sp500_df["SPY"] = spy_df["close"]
close_sp500_df.index = close_sp500_df.index.date

daily_returns_sp500 = close_sp500_df.pct_change().dropna()
daily_returns_sp500.head()
```

```python
close_agg_df = pd.DataFrame()
close_agg_df["AGG"] = agg_df["close"]
close_agg_df.index = close_agg_df.index.date

daily_returns_agg = close_agg_df.pct_change().dropna()
daily_returns_agg.head()
```

```python
close_gld_df = pd.DataFrame()
close_gld_df["GLD"] = gld_df["close"]
close_gld_df.index = close_gld_df.index.date

daily_returns_gld = close_gld_df.pct_change().dropna()
daily_returns_gld.head()
```

```python
close_usrt_df = pd.DataFrame()
close_usrt_df["USRT"] = usrt_df["close"]
close_usrt_df.index = close_usrt_df.index.date

daily_returns_usrt = close_usrt_df.pct_change().dropna()
daily_returns_usrt.head()
```

```python
# Concatenate into a Dataframe
compare_df = pd.concat([daily_returns_sp500,daily_returns_agg,daily_returns_gld,daily_returns_usrt],axis=1)
compare_df
```

```python
# Plot
```

```python
compare_df.plot(xlabel="Dates",
    ylabel="Daily Returns",
    rot=45
)
```

```python
# Hvplot
```

```python
compare_daily_returns_df = compare_df.hvplot.line(
    xlabel="Year",
    ylabel="Daily Returns",
    title='Daily Returns for each ETF')

compare_daily_returns_df
```

```python
# Monte Carlo Simulation Before Covid-19
```

```python
tickers = ["SPY","AGG","GLD","USRT"]

start_date_mc = pd.Timestamp("2019-03-01", tz="America/New_York").isoformat()
end_date_mc = pd.Timestamp("2020-03-01", tz="America/New_York").isoformat()

prices_df = alpaca.get_bars(
    tickers,
    timeframe,
    start= start_date_mc,
    end= end_date_mc).df

AGG = prices_df[prices_df['symbol']=='AGG'].drop('symbol', axis=1)
SPY = prices_df[prices_df['symbol']=='SPY'].drop('symbol', axis=1)
GLD = prices_df[prices_df['symbol']=='GLD'].drop('symbol', axis=1)
USRT = prices_df[prices_df['symbol']=='USRT'].drop('symbol', axis=1)

prices_df = pd.concat([SPY,AGG,GLD,USRT],axis=1, keys=['SPY','AGG','GLD','USRT'])
```

```python
monte_carlo_simulation = MCSimulation(
    portfolio_data= prices_df,
    weights=[.25,.25,.25,.25],
    num_simulation= 10,
    num_trading_days= 252 * 10)

monte_carlo_simulation.portfolio_data.head()
```

```python
monte_carlo_simulation.calc_cumulative_return()
```

```python
mc_line_plot = monte_carlo_simulation.plot_simulation()
```
